```python
fruits = ['apple', 'oranges', 'grapes']
for f in fruits:
  print(f)
```

```python
it = iter(fruits)
print(next(it))
print(next(it))
print(next(it))
```

```python
def is_iterable(obj):
  try:
    iter(obj)
    return True
  except TypeError:
    return False

print(is_iterable(fruits))
print(is_iterable(10))
```
